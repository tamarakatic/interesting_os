```python
import matplotlib as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```python
order = ["(m, j, k)", "(j, m, k)", "(m, k, j)", "(j, k, m)", "(k, m, j)", "(k, j, m)"]
```

```python
n_1000 = [[4.866718, 4.277955, 3.332042, 7.417254, 3.472387, 7.245092]]
n_2000 = [[63.641975, 50.594655, 26.381583, 76.018497, 26.718034, 76.419062]]
n_5000 = [[1113.917325, 902.709071, 415.012441, 1347.945309, 415.224646, 1356.065067]]
```

```python
df_1000 = pd.DataFrame(n_1000, columns=order, index={0: "1000"})
df_2000 = pd.DataFrame(n_2000, columns=order, index={0: "2000"})
df_5000 = pd.DataFrame(n_5000, columns=order, index={0: "5000"})
```

```python
df_1000.rename(index={0: "1000"}, inplace=True)
df_2000.rename(index={0: "2000"}, inplace=True)
df_5000.rename(index={0: "5000"}, inplace=True)
```

```python
df_1000.plot(kind="bar", figsize=(8, 6), alpha=0.75, rot=0)
df_2000.plot(kind="bar", figsize=(8, 6), alpha=0.75, rot=0)
df_5000.plot(kind="bar", figsize=(8, 6), alpha=0.75, rot=0)
```

```python
new = n_1000 + n_2000 + n_5000
```

```python
df = pd.DataFrame(new, columns=order)
```

```python
df.rename(index={0: "1000", 1: "2000", 2: "5000"}, inplace=True)
```
